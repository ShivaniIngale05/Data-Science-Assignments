```python
# Ranking-based metrics
def hit_ratio_at_k(y_true, y_scores, k=5):
    top_k_idx = np.argsort(y_scores)[-k:][::-1]
    return int(any(np.array(y_true)[top_k_idx]))

def ndcg_at_k(y_true, y_scores, k=5):
    order = np.argsort(y_scores)[::-1]
    y_true_sorted = np.array(y_true)[order[:k]]
    dcg = np.sum(y_true_sorted / np.log2(np.arange(2, len(y_true_sorted) + 2)))
    ideal_sorted = np.sort(y_true)[::-1][:k]
    idcg = np.sum(ideal_sorted / np.log2(np.arange(2, len(ideal_sorted) + 2)))
    return dcg / idcg if idcg > 0 else 0

print("\nðŸ”¹ Hit Ratio@5:", hit_ratio_at_k(y_true, y_scores, k=5))
print("ðŸ”¹ NDCG@5:", round(ndcg_at_k(y_true, y_scores, k=5), 3))
```

## Interview Questions

#### 1. Can you explain the difference between user-based and item-based collaborative filtering?

User-based collaborative filtering recommends items by finding similar users to the target user and then suggesting items those similar users liked. In contrast, Item-based collaborative filtering recommends items by finding items similar to those the target user has already liked, based on relationships between items themselves rather than user similarities.

User-based methods are more personalized but struggle with scalability and "cold-start" (new users/items), while item-based methods scale better and handle new items more gracefully but can offer less personalizedÂ results.

#### 2. What is collaborative filtering, and how does it work?

Collaborative filtering is a machine learning method used by recommender systems to predict user preferences by finding patterns in the behavior of many users to suggest itemsÂ orÂ content.

It works on the principle that if users have similar tastes in the past, they will continue to agree on future preferences, recommending items to one user based on what other similar users have liked. This technique is widely used by platforms like Netflix and Amazon to personalize the user experience by suggesting relevant contentÂ andÂ products.
