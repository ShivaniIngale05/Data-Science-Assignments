Scaling techniques are essential in machine learning to ensure that features are on the same scale, which can improve model performance and convergence.


Scaling Techniques:
1. Standard Scaling (Z-score normalization): Scales data to have a mean of 0 and a standard deviation of 1.
2. Min-Max Scaling: Scales data to a fixed range, typically between 0 and 1.
3. Robust Scaling: Uses median and interquartile range for scaling, making it robust to outliers.


Scenarios for Each Scaling Technique:
1. Standard Scaling:
- Preferred when: Features are approximately normally distributed and algorithms like linear regression, logistic regression, or neural networks are used.
- Why: Many algorithms (like those using gradient descent) perform better when features are centered around zero with similar scales.
- Examples:
    - Linear Regression
    - Logistic Regression
    - Neural Networks (often benefit from standardized inputs)


2. Min-Max Scaling:
- Preferred when: Algorithms require bounded input ranges (like some neural networks) or distance calculations are involved (like k-NN).
- Why: Scales features to a specific range (often [0,1]), useful for algorithms sensitive to scale like neural networks with certain activation functions.
- Examples:
    - Neural Networks with sigmoid activation functions
    - k-Nearest Neighbors (k-NN)
    - Image processing (pixel values often scaled to [0,1])


. Robust Scaling:
- Preferred when: Data contains outliers and you want scaling robust to them.
- Why: Uses median and interquartile range, making it less sensitive to outliers compared to Standard Scaling.
- Examples:
    - Datasets with significant outliers
    - When data distribution isn't Gaussian and contains extreme values

##### One Hot Encoding :

Definition: Creates a new binary column for each category. For example, sex = {Male, Female} becomes sex_Male = 1/0, sex_Female = 1/0.

Pros:

No ordinal assumption → Works well for nominal categories (where no natural order exists, e.g., colors, gender).

Prevents misinterpretation → Models don’t mistakenly assume one category is larger or smaller than another.

Widely supported → Works well with most ML algorithms (linear regression, logistic regression, decision trees, etc.).


Cons:

High dimensionality → If a feature has many categories, it creates lots of new columns (sparse matrix → more memory and computation).

Curse of dimensionality → Too many dummy variables can reduce model performance.

Not suitable for tree-based models in large datasets (they handle label encoding better).

##### Label Encoding :

Definition: Assigns a unique integer value to each category (e.g., Red=0, Green=1, Blue=2).

Pros:

Memory efficient → Only one column per feature.

Fast to implement → Simple transformation using LabelEncoder.

Works well for tree-based models (Decision Trees, Random Forest, XGBoost) since they split on values, not magnitude.

Cons:

Imposes ordinality → Creates a false assumption of order (e.g., Blue=2 > Red=0), which can mislead linear/logistic regression, k-NN, or SVM.

Not suitable for nominal data when there is no natural ranking.

Can cause bias in distance-based algorithms (k-means, k-NN) since encoded values affect distance calculation.

## Feature Engineering

```python
# 1. Feature: Work Hours Category
def work_hours_type(hours):
    if hours < 30:
        return "Part-time"
    elif 30 <= hours <= 40:
        return "Full-time"
    else:
        return "Over-time"

df["work_hours_type"] = df["hours_per_week"].apply(work_hours_type)
```

```python
# 2. Feature: Age Group
df["age_group"] = pd.cut(df["age"],
                         bins=[0, 25, 45, 65, 100],
                         labels=["Young", "Mid-age", "Senior", "Elder"])
```

```python
# Show updated dataset
print(df[["age", "age_group", "hours_per_week", "work_hours_type"]].head(10))
```

1. Work Hours Category (work_hours_type)

Why created?
Raw hours_per_week is continuous, but what really matters for income prediction is the type of work schedule.

People working part-time (<30 hrs) usually earn less.

Those working full-time (30–40 hrs) have stable salaries.

Those working overtime (>40 hrs) often belong to higher-paying jobs or are more committed, which may correlate with >50K income.


Benefit to the model :

Instead of letting the model guess patterns from raw hours, categorizing into Part-time / Full-time / Over-time captures meaningful human distinctions, improving interpretability and predictive power.


2. Age Group (age_group)

Why created?
Income is not linearly related to age.

Young (<25 yrs) → Less work experience, usually lower income.

Mid-age (25–45 yrs) → Prime working years, higher chance of earning >50K.

Senior (45–65 yrs) → Experienced, but income growth slows down.

Elder (>65 yrs) → Many may retire or work fewer hours, affecting income.


Benefit to the model :

Instead of treating every age as a separate value, grouping captures life stage effects on income, reducing noise and making the feature more informative.

Why capital_gain (and capital_loss)?

These features are highly skewed:

Most values are 0.

A few people have very large amounts (outliers).


This creates a right-skewed distribution, which can distort model training:

Linear models may place too much weight on a few extreme values.

Distance-based models (k-NN, clustering) may treat outliers as very different, even if they aren’t.


Why Log Transformation?

Reduces skewness → Log transformation compresses large values, making the distribution more balanced.

Preserves order → Higher values remain higher, but the difference between extreme values is reduced.

Improves model stability → Models that assume normally distributed features (e.g., linear/logistic regression) perform better when input distributions are less skewed.

Helps visualization → Makes it easier to detect patterns and relationships with other features.


 ##### We chose log transformation for capital_gain because it is highly skewed with extreme outliers. The log transform compresses these large values, reduces skewness, and ensures that models can learn general patterns rather than being dominated by a few extreme cases.


## Feature Selection

```python
from sklearn.ensemble import IsolationForest
```

```python
# Select numerical features only (outlier detection works best on numerical data)
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
num_cols
```

```python
# Isolation Forest model
iso = IsolationForest(contamination=0.02, random_state=42)  # contamination = % of outliers expected
iso
```

```python
y_pred = iso.fit_predict(df[num_cols])
y_pred
```

```python
# -1 = outlier, 1 = normal
df['outlier'] = y_pred
```

```python
# Remove outliers
df_cleaned = df[df['outlier'] == 1].drop(columns=['outlier'])
df_cleaned
```

```python
print("Original dataset shape:", df.shape)
print("Cleaned dataset shape:", df_cleaned.shape)
```

1. Bias the model: Outliers can pull regression lines or decision boundaries in the wrong direction.

Example: In linear regression, a single extreme value can drastically change the slope.



2. Increase variance: Models like k-NN, clustering, and SVM can give undue importance to distant points, reducing generalization.


3. Hurt accuracy: Outliers add noise and make the model learn patterns that don’t represent the majority.


4. Slow training: Extreme values may increase computation or force the model to use more complex decision boundaries.


Benefit of Removing Outliers:

Improves generalization by making patterns more representative.

Prevents extreme cases from dominating model training.

Leads to better accuracy and stability.
