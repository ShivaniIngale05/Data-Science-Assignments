### Exercise 1: Prime Numbers
##### Write a Python program that checks whether a given number is prime or not. A prime number is a natural number greater than 1 that has no positive divisors other than 1 and itself.

```python
def is_prime(n):   
    if n <= 1:      
        return False
    for i in range(2, n):     
        if n % i == 0:
            return False
    return True

num = int(input("Enter a number: "))
if is_prime(num):
    print(f"{num} is a prime number.")
else:
    print(f"{num} is not a prime number.")
```

### Exercise 2: Product of Random Numbers
##### Q. Develop a Python program that generates two random numbers and asks the user to enter the product of these numbers. The program should then check if the user's answer is correct and display an appropriate message.

```python
import random

num1 = random.randint(1, 10)     # Generate two random numbers num1 and num2
num2 = random.randint(1, 10)

print(f"What is {num1} * {num2}?")

user_answer = int(input("Enter your answer: "))   # user enter the product

if user_answer == num1 * num2:   # check if answer is correct
    print("Correct! Well done.")
else:
    print(f"Incorrect. The correct answer is {num1 * num2}.")
```

### Exercise 3: Squares of Even/Odd Numbers
##### Create a Python script that prints the squares of all even or odd numbers within the range of 100 to 200. Choose either even or odd numbers and document your choice in the code.


```python
# for even numbers
for num in range(100, 201):  # give the range between 100-200
    if num % 2 == 0:         # if the number is divided by 2 then it is even 
        print(f"{num}^2= {num**2}")   # taking the square of num
```

```python
# for odd numbers
for num in range(100, 201):  
    if num % 2 != 0:     # # if the number is not  divided by 2 then it is odd 
        print(f"{num}^2= {num**2}")     # taking the square of num
```

### Exercise 4: Word counter
##### write a program to count the number of words in a given text.

```python
text = "This is a sample text. This text will be used to demonstrate the word counter."
words = text.split()     # Split text into words based on spaces
word_count = {}          # create empty dictionary to store word count

for word in words:           # Count occurrences of each word
    word_count[word] = word_count.get(word, 0) + 1
    
for word, count in word_count.items():    # result
    print(f"'{word}': {count}")
```

### Exercise 5: Check for Palindrome
##### Write a Python function called is_palindrome that takes a string as input and returns True if the string is a palindrome, and False otherwise. A palindrome is a word, phrase, number, or other sequence of characters that reads the same forward and backward, ignoring spaces, punctuation, and capitalization.

```python
import re

def is_palindrome(text):
    text = text.lower()  # convert text into lower case
    cleaned = re.sub(r'[^a-z0-9]', '', text) # keep only alphanumeric characters
    return cleaned == cleaned[::-1] #  Check if string equals its reverse
```

```python
is_palindrome("racecar")      
```

```python
is_palindrome("Hello")
```
