## 1. Using Sample Standard Deviation

```python
import numpy as np
import scipy.stats as stats
import statistics
```

```python
#sample data
data=[1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29]
data
```

```python
#length of sample data
n=len(data)  
n
```

```python
mean=np.mean(data)   #sample mean
mean
```

```python
sd=np.std(data)   #sample standard deviation
sd
```

```python
df=n-1   #degree of freedom
df
```

```python
# t-critical value for 99% confidence interval

alpha = 0.01
t_critical =stats.t.ppf(1-alpha/2,df)
t_critical
```

```python
# margin of error
margin_error = t_critical * (sd / np.sqrt(n))
margin_error
```

```python
# confidence interval(lower)

CI_lower_bound = mean - margin_error
CI_lower_bound
```

```python
# # confidence interval(upper)

CI_upper_bound = mean + margin_error
CI_upper_bound
```

### Interpretation

we use t-distribution because the population standard deviation is unknown and also sample size is small.

using a t-distribution, we are 99% confident  that the true mean number of charecters printed before the print-head fails lies between the 1.095231668626123 and 1.3821016647072102.

that means if we took many similar samples of size 15 and build a 99% confidence interval each time, about 99% of those intervals would contain the true population mean.
