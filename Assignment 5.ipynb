```python
# Check missing values
print("\nMissing Values per Column:")
print(df.isnull().sum())
```

```python
# Identify & Correct Data Type Inconsistencies
print("\n--- Original Data Types ---")
print(df.dtypes)
```

```python
# Try converting object columns to numeric (if possible)
for col in df.columns:
    if df[col].dtype == 'object':
        try:
            df[col] = pd.to_numeric(df[col])  # convert string numbers -> numeric
            print(f"Converted {col} to numeric")
        except ValueError:
            # If conversion fails, it's categorical
            df[col] = df[col].astype('category')
            print(f"Converted {col} to category")
```

```python
# Ensure categorical variables use 'category' dtype
categorical_cols = df.select_dtypes(include=['object']).columns
for col in categorical_cols:
    df[col] = df[col].astype('category')
    print(f"Converted {col} to category")

# 3. Verify corrected data types
print("\n--- Updated Data Types ---")
print(df.dtypes)
```

```python
for col in df.select_dtypes(include=['number']):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outlier_count = ((df[col] < lower) | (df[col] > upper)).sum()
    print(f"{col}: {outlier_count}Â outliers")
```

## Statistical Summary

```python
# Statistical Summary for Each Variable
def summary_statistics(df):
    stats_df = pd.DataFrame({
        "Mean": df.mean(numeric_only=True),
        "Median": df.median(numeric_only=True),
        "Std Dev": df.std(numeric_only=True),
        "IQR": df.quantile(0.75, numeric_only=True) - df.quantile(0.25, numeric_only=True)
    })
    return stats_df

stats_summary = summary_statistics(df)

print("\n--- Statistical Summary ---")
display(stats_summary)
```

```python
# 2. Bar charts for categorical features
categorical_cols = df.select_dtypes(include=['object','category']).columns
for col in categorical_cols:
    plt.figure(figsize=(6, 4))
    sns.countplot(x=col, data=df)
    plt.title(f"Bar Chart of {col}")
    plt.show()
```

```python
# Scatter plot example (LB vs AC if available)
if "LB" in df.columns and "AC" in df.columns:
    plt.figure(figsize=(6, 4))
    sns.scatterplot(x="LB", y="AC", data=df)
    plt.title("Scatter Plot: LB vs AC")
    plt.show()
```

```python
# Pair plot for deeper relationship exploration
sns.pairplot(df, diag_kind="kde")
plt.suptitle("Pair Plot of Numerical Variables", y=1)
plt.show()
```

## 4.Pattern Recognition and Insights

```python
# Pattern Recogniti
# 1. Correlation Analysis
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="viridis", fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

```python
# 2. Trends over time (if dataset has a time column like 'Date')
if "Date" in df.columns:
    df["Date"] = pd.to_datetime(df["Date"], errors="coerce")
    df.set_index("Date", inplace=True)
    df.resample("M").mean().plot(figsize=(12,6))
    plt.title("Monthly Trends of Numerical Variables")
    plt.show()
```
