## INTERPRETATION

1. Average Price

The mean and median values are relatively close, suggesting that product prices are fairly balanced without extreme skewness.
A moderate standard deviation indicates that while most products fall in a common price range, there are a few high-priced items that increase variability.
The mode reflects the most common product price point in the dataset.

2. Total Sales Value

The high standard deviation shows a wide spread in sales values, meaning that some products or transactions generate significantly higher sales compared to others.
The mean being much higher than the median suggests that a few very large sales transactions are pulling the average up, indicating right-skewed distribution.

3. Discount Rate (%)

The average discount percentage shows the general level of promotions offered.
The mode indicates the most common discount rate given, which could represent a standard promotion strategy (e.g., 10% or 15%).
A moderate spread suggests that discounts vary but are not extremely inconsistent.

4. Discount Amount

The high variability in discount amounts reflects that some transactions involved large price reductions (likely due to high-value items), while others involved smaller discounts.
Median discount amounts provide a more typical view of what most customers received, since the mean may be inflated by a few large discounts.

5. Net Sales Value

The average net sales value shows the actual revenue after discounts.
Since the mean is higher than the median, the data is skewed by a few very large transactions.
A large standard deviation confirms that sales performance is uneven across products or categories.

```python
for col in df.select_dtypes(include=['number']):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outlier_count = ((df[col] < lower) | (df[col] > upper)).sum()
    print(f"{col}: {outlier_count} outliers")
```

## Standardization of Numerical Variables

```python
num_cols = ["Volume", "Avg Price", "Total Sales Value", 
            "Discount Rate (%)", "Discount Amount", "Net Sales Value"]
num_cols
```

```python
df_standardized = (df[num_cols] - means) / std
df_standardized
```

```python
df_compare = pd.concat([df[num_cols], df_standardized], axis=1)
df_compare.columns = [f"{col} (Original)" for col in num_cols] + [f"{col} (Standardized)" for col in num_cols]
df_compare
```

```python
df_standardized.mean()
```

```python
 df_standardized.std()
```

```python
cols_to_plot = ["Volume", "Avg Price", "Net Sales Value"]
plt.figure(figsize=(15, 10))

for i, col in enumerate(cols_to_plot):
    # Original distribution
    plt.subplot(2, len(cols_to_plot), i+1)
    plt.hist(df[col], bins=20, edgecolor="black")
    plt.title(f"{col} (Original)")
    
    # Standardized distribution
    plt.subplot(2, len(cols_to_plot), i+1+len(cols_to_plot))
    plt.hist(df_standardized[col], bins=20, edgecolor="black")
    plt.title(f"{col} (Standardized)")

plt.tight_layout()
plt.show()
```

### Conclusion

After applying z-score standardization, all numerical variables in the dataset were transformed to have:
Mean = 0 and Standard deviation = 1

After standardization, all numerical variables are rescaled to have mean 0 and standard deviation 1, preserving their distribution shape but removing differences in scale.
